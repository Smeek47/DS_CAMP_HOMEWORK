```python
import warnings
# Turning warnings off
warnings.filterwarnings("ignore")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
# %matplotlib notebook
```

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=5,
                  cluster_std=0.6, random_state=0)

colors = np.array([plt.cm.Spectral(val)
                   for val in np.linspace(0, 1, len(set(y)))])

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors[y], s=20)
```

### Agglomerative clustering using "complete" linkage 


```python
from sklearn.cluster import AgglomerativeClustering


clt = AgglomerativeClustering(n_clusters=5,
                              linkage="complete",
                              )

labels = clt.fit_predict(X, y)
plt.scatter(X[:, 0], X[:, 1], c=colors[labels], s=20, edgecolor='black')
```

### Agglomerative clustering using "single" linkage 

```python
clt = AgglomerativeClustering(n_clusters=5,
                              linkage="single",
                              )

labels = clt.fit_predict(X, y)
plt.scatter(X[:, 0], X[:, 1], c=colors[labels], s=20, edgecolor='black')
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN


clt = DBSCAN(eps=0.56,
             min_samples=5
             )

labels = clt.fit_predict(X, y)
plt.scatter(X[:, 0], X[:, 1], c=colors[labels], s=20, edgecolor='black')
```

### Use elbow method to determine the best k for Kmeans

```python
from sklearn.cluster import KMeans


number_of_clusters = range(1, 8)
wcss = []

for k in number_of_clusters:
    clt = KMeans(n_clusters=k)
    clt.fit(X)
    wcss.append(clt.inertia_)

plt.plot(number_of_clusters, wcss, lw=5)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
plt.show()
```

# The optimal number of clusters for this case is <font color="lime">3</font>.

```python
km = KMeans(3)
km.fit(X)
labels = km.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=colors[labels], s=20, edgecolor='black')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], c='b', marker='x', s=300, lw=3)
```
