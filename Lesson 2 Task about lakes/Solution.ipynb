# Input data


## Task about lakes

- Generate 20 random variable from 1 to 100
- Draw the plot of the  sequence 
- Lets assume each point represent the height and so all plotting is 2d mountains. 
- Then consider the unlimited rain from above - cavities become lakes full of water. 
- Determine the deepest lake


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2023)
a = np.random.randint(1,100,20)
print(a)
plt.plot(a)
```

# Solution

<hr/>

```python
def find_deepest_lake(a):
    deepest_lake = np.argmin(a) # Знайшли найглибшу точку озера.
    # Тепер знайдемо найбільші точки по ліву і по праву сторону найглибшої точки.
    top_left, top_right = np.argmax(a[:deepest_lake]), np.argmax(a[deepest_lake:]) + deepest_lake   # Щоб не втратити послідовність нумерації
                                                                                                    # індексів додамо deepest_lake
    plt.plot(a)
    plt.plot(range(top_left, top_right + 1), 
            [a[i] for i in range(top_left, top_right + 1)],
            c='red',)

find_deepest_lake(a)
```
