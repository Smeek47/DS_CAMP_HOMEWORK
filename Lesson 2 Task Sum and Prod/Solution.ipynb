```
There are two whole numbers:
1 < a,b <100

One scientist("Sum") get provided with sum of numbers, ( S )
another  ("Prod") get provided with product of numbers. ( P )
Both scientists know that numbers 1 < a,b <100.

Determine the numbers being based on the following dialog: 
    Prod: I don't know the numbers;
    Sum: I know it;
    Prod: then I know the numbers; 
    Sum: then I know the numbers too.
```
a + b = S<br>
a * b = P
<hr/>


До діалогу науковців значення P  може бути будь-яким добутком a i b в межах 1<a,b<100

````python
possible_P_1 = {} # Початкові можливі значеня P

for a in range(2, 100): 
    for b in range(a, 100): 
        if a*b not in possible_P_1: 
            possible_P_1[a*b] = 1 
        else:
            possible_P_1[a*b] += 1 # Одразу записуємо кількість пар чисел з яких можна скласти "i" число

print(possible_P_1)
````

    Prod: I don't know the numbers;
 Ця фраза означає, що у цього науковця число (Р), яке можна скласти мінімум з 2 різних пар множників (не унікальне).<br/>
 Отже сума (S) не може бути складена з чисел а+b які при множенні дадуть унікальний добуток.
 

````python
cannot_be_S = {}  # Числа які не можуть бути S
for a in range(2, 100): 
    for b in range(a, 100): 
        if  possible_P_1[a*b] == 1 :
            cannot_be_S[a+b] = 1  

print(cannot_be_S)
````

    Sum: I know it;
Науковець Sum так сказав знаючи, що Prod не визначить число.<br>
Але науковець Prod на основі цієї репліки зробив свої висновки. <br>
Він звузив діапазон можливих добутків(P possible) розкладаючи на доданки можливі суми, і порівнювати чи є серед добутків доданків такі, які складаються більш ніж з 1 пари множників.


````python
possible_P_2 = {} # звужений діапазон значень Р
for n in range(2, 100):
    if n not in cannot_be_S:
        for a in range(2, (n//2)+1):
            p = a * (n-a)
            if p in possible_P_1 and possible_P_1[p] > 1:
                if p in possible_P_2:
                    possible_P_2[p] += 1
                else:
                    possible_P_2[p] = 1 

print(possible_P_2)
````

    Prod: then I know the numbers; 
Можливі лише ті P, які можуть бути поділені на a та b,<br/>і де а + b відбувається тільки 1 раз (possible_P_2[P] == 1)

````python
possible_S = {}  # фільтруємо можливі варіанти S 
for n in range(2, 100):
    if n not in cannot_be_S:
        for a in range(2, (n//2)+1):
            if a*(n-a) in possible_P_2 and possible_P_2[a*(n-a)] == 1:
                if n in possible_S:
                    possible_S[n] += 1
                else:
                    possible_S[n] = 1
print(possible_S)
````

    Sum: then I know the numbers too.
Оскільки Sum тепер знає відповідь, то поділ на доданки відбудеться тільки коли possible_S[S]==1

````python
for n in possible_S: 
    if possible_S[n] == 1:
        for a in range(2, (n//2)+1):
            if a*(n-a) in possible_P_2 and possible_P_2[a*(n-a)] == 1:
                print (f"S = {n}; P = {a*(n-a)}; a = {a}; b = {n-a};")
````
