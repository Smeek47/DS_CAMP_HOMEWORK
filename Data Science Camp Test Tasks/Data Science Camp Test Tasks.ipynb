```python
import numpy as np


def solve_task1(A:list, B:list):
    X = None
    try:
        X = np.linalg.solve(A, B)
        print(" :")
        print("X =", np.transpose(X))
    except np.linalg.LinAlgError:
        print("     ' (  A).")
    except Exception as e:
        print(f"  : {e}")
    return X

def manual_filling():

    for i in range(3):
        A[i] = [float(input(f" A[{i + 1},{j + 1}]: ")) for j in range(3)]

    for i in range(3):
        B[i] = float(input(f" B[{i + 1}]: "))

    return A,B


A = np.array([[1, 2, 3],
              [0, 1, 2],
              [2, 0, 0]])

B = np.array([[1],
              [1],
              [0]])

#  

solve_task1(A,B)

#  ,      A  B

manual_a,manual_b = manual_filling()
solve_task1(manual_a,manual_b)


```

```python

def probability_of_drop(prob_side_drop:list,
                        prob_coin_chosen:list):
    #        
    #     :
    # prob_side_drop -        
    # prob_coin_chosen -      
    p = 0
    for i in range(len(prob_side_drop)):
        p += float(prob_side_drop[i])*float(prob_coin_chosen[i])

    return p

def coin_prob_chosen(prob_side_drop:list,
                     previous_prob_coin_chosen:list,
                     previous_probability_of_drop:float):
    #        
    #         
    p = []
    for i in range(len(prob_side_drop)):
        p.append(round((float(prob_side_drop[i])*float(previous_prob_coin_chosen[i]))  /  float(previous_probability_of_drop),5))
    return p

h_drop_chanse = [(0.1),(0.2),(0.4),(0.8),(0.9)]

history = ["H","H","H","T","H","T","H","H"]

#         . 
#          - 1/5
coins_probability_chosen = [([1/len(h_drop_chanse) for i in range(len(h_drop_chanse))])]

answer = []

for i in range(len(history)):

    # ---[        ,      ]---
    if i == 0 and history[0]== "H":
        answer.append(
            round(
                probability_of_drop(h_drop_chanse,coins_probability_chosen[i])
                ,2
            )
        )
        continue
    elif i == 0 and history[0] == "T" :
        answer.append(
            round(
                probability_of_drop([1-h_drop_chanse[_] for _ in range(len(h_drop_chanse))],coins_probability_chosen[i])
                ,2
            )
        )
        continue
    # --------------

    # -----[         ]---------
    if history[i-1] == "H":
        coins_probability_chosen.append(
            coin_prob_chosen(h_drop_chanse,
                             coins_probability_chosen[i-1],
                             answer[i-1])
        )
    elif history[i-1] == "T":
        coins_probability_chosen.append(
            coin_prob_chosen([1-h_drop_chanse[_] for _ in range(len(h_drop_chanse))],
                             coins_probability_chosen[i-1],
                             1-answer[i-1])
        )
    # -----------------------------------------------------------------------------

    # --------[     ]-------------------
    answer.append(
    round(
        probability_of_drop(h_drop_chanse,coins_probability_chosen[i])
        ,2
    )
    )
    # -----------------------------------------------------------



#   
print("C    :")
print(history)
print(answer)

element9 = round(probability_of_drop(h_drop_chanse,coin_prob_chosen(h_drop_chanse,coins_probability_chosen[7],answer[7])),2)
print(f"  '   H = {element9},  T = {round(1-element9,2)} ")



#                               
#    ,        1  100%
# (        1,      . )
#   ,      

# print(" ")
# previrka=[]
# for i in range(len(coins_probability_chosen)):
#     previrka.append(round(sum(coins_probability_chosen[i]),2))
# print(previrka)
```
